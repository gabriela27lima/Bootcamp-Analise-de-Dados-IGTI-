```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clientes = pd.read_csv('clientes_8000.csv', sep =';')
```

```python
#não sei como funciona mas é importante para colocar o gráfico em ordem decrescente
res = (
clientes
.groupby('estado_civil')
.salario
.mean()
.sort_values(ascending=False)
)
```

```python
fig,grafico_1 = plt.subplots(figsize=(8,5)) 
sns.barplot(x=clientes.estado_civil, y=clientes.salario, palette='RdPu', order = res.index)
grafico_1.set_title("Salário X Estado Civil", fontdict={'fontsize':18})
grafico_1.set_xlabel('Estado Civil', fontdict={'fontsize':16})
grafico_1.set_ylabel('Média Salarial', fontdict={'fontsize':16})
grafico_1.spines['bottom'].set_linewidth(2.5)
for axis in ['top', 'right', 'left']:
   grafico_1.spines[axis].set_color(None)
fig.tight_layout()
```
