```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clientes = pd.read_csv('clientes_8000.csv', sep =';')
```

```python
#Média das variáveis quantitativas (total do dataframe) por determinado grupo, neste caso, por estado civil e sexo.
clientes.groupby(['estado_civil', 'sexo']).agg(np.mean)
```

```python
#Criando objeto com pandas (o que?) - Média salarial por estado civil e sexo
_table = pd.crosstab(clientes.estado_civil, clientes.sexo, aggfunc = 'mean', values = clientes.salario)
```

```python
#Apresentando por níveis de gradiente
_table.style.background_gradient()
```

```python
#Bloxplot (Não acho esse bonito)
bp_salario_sexo = sns.boxplot(x = 'salario', y= 'sexo', data = clientes, orient = 'h')
bp_salario_sexo.set_title('Renda', fontsize = 18)
bp_salario_sexo.set_xlabel('Salário(R$)', fontsize = 14)
bp_salario_sexo.set_ylabel('Sexo', fontsize = 14)
```
