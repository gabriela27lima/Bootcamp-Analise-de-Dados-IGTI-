```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clientes = pd.read_csv('clientes_8000.csv', sep =';')
```

```python
clientes.estado_civil #coluna Estado Civil
```

```python
#Frequencia absoluta
clientes['estado_civil'].value_counts()
```

```python
#Frequencia percentual
clientes['estado_civil'].value_counts(normalize = True)*100
```

```python
#criando objeto para frequencia absoluta
fa_estado_civil=clientes['estado_civil'].value_counts()
fa_estado_civil
```

```python
#plotando a frequência absoluta
fa_estado_civil.plot()
```

```python
#criando objeto para frequencia percentual
fp_estado_civil=clientes['estado_civil'].value_counts(normalize = True)*100
fp_estado_civil
```

```python
#Criando DataFrame para Frequencias absolutas e percentuais
fq_estado_civil = pd.DataFrame({'Frequência': fa_estado_civil, 'Porcentagem(%)': fp_estado_civil})
fq_estado_civil
```

```python
#Introduzindo nome na primeira coluna para descrição
fq_estado_civil.rename_axis('Estado Civil', axis = 'columns', inplace = True)
fq_estado_civil
```
